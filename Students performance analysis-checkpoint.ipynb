# Student Performance Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = pd.read_csv("StudentsData.csv")
a
```

```python
# Q1. How can we check basic information and missing values in the dataset?
a.info()
a.describe()
a.isnull().sum()
```

```python
# Q2. How to analyze gender-wise performance in Math?
sns.barplot(x='gender', y='math score', data=a)
plt.title("Gender-wise Math Performance")
plt.show()
```

```python
# Q3. How to find the correlation between Math and Reading scores?
sns.scatterplot(x='math score', y='reading score', data=a)
plt.title("Math vs Reading Score Correlation")
plt.show()

a[['math score','reading score']].corr()
```

```python
# Q4. How to analyze the effect of study hours on performance?
import numpy as np
a['study_hours'] = np.random.randint(1, 5, size=len(a))

sns.barplot(x='study_hours', y='math score', data=a)
plt.title("Study Hours vs Math Score")
plt.show()
```

```python
# Q5. How to visualize gender-wise writing score distribution?
plt.figure(figsize=(8,5))
sns.boxplot(x='gender', y='writing score', data=a)
plt.title("Gender-wise Writing Score Distribution")
plt.show()
```

```python
# Q6. How to calculate and visualize each studentâ€™s average score?
a['average_score'] = a[['math score','reading score','writing score']].mean(axis=1)
sns.histplot(a['average_score'], bins=10, kde=True)
plt.title("Average Score Distribution")
plt.show()
```
